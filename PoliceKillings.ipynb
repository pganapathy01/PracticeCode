```python
# %load PoliceKillings.py
__author__ = 'Priya'

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import operator
```

```python
readfile = pd.read_csv('PoliceKillings.csv')
print(readfile)

f = readfile["year"]

print(f)


rowsWithStates = readfile["state"]

print(rowsWithStates)


State = []
for row in rowsWithStates:
    if row not in State:
       State.append(row)
print(State)

killingsbystate ={}
populationbystate={}

for row in State:
         killingsbystate[row] = sum(readfile["state"]==row)
         populationbystate[row]=sum(readfile["pop"][readfile["state"]==row])
print(killingsbystate)
print(populationbystate)

# Normalize killings per state
normalize_killings_state ={}
for key,value in killingsbystate.iteritems():
    normalize_killings_state[key] =np.float64(killingsbystate[key]/populationbystate[key])*100
print(normalize_killings_state)


# Find Top 3 highest cop killing states
killingsbystate= sorted(killingsbystate.items(), key=operator.itemgetter(1),reverse=True)
populationbystate= sorted(populationbystate.items(), key=operator.itemgetter(1),reverse=True)
```
